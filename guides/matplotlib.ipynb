**Well, that didn't work. I searched around for why, from this [GitHub issue](https://github.com/jupyter/notebook/issues/3523): "Splitting matplotlib statements into multiple Jupyter input cells may cause nothing to appear. Check if all your statements are in the same cell."**

```python
# All together:
fig,ax = plt.subplots()

x_values = ['9am', '10am', '11am', '12pm', '1pm', '2pm', '3pm', '4pm', '5pm']

y_values = [367.22, 400, 120, 386, 365, 379, 350, 400, 408.85]

assert len(x_values) == len(y_values)
ax.plot(x_values, y_values, color="blue", marker="s", label="Blue Line Jumps 11 Percent" )
ax.legend()
plt.show()
```
